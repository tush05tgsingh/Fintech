## Where does the Financial data come from 

1) Market data - stock prices, exchange rates, crypto. Interest rates
2) Fundamental Data: company balance sheets, income statements, ratios.
3) Alternative Data: news sentiment, social media, satellite images (yep, hedge funds use these!).
4) High-frequency vs. low-frequency: tick-level vs. daily/quarterly.

```python
import yfinance as yf

data = yf.Ticker("AAPL")
hist = data.history(period="1mo") #price data for the last month
fundamentals = data.financials #company financial statements
print(hist.head())
print(fundamentals.head())
```

Let's compare the price data using charts and tables depending on the data 

```python
## Price data is time series data, so we visualise it using line charts
import matplotlib.pyplot as plt
hist['Close'].plot(title="AAPL Stock Price")
plt.show()
```

```python
## Financial statements are tabular data, so we visualise it using tables
#print(fundamentals)

# you can also show the data in a more readable format using pandas
import pandas as pd
data = pd.DataFrame(fundamentals.T)
print(data)

```

1. Market data = the “heartbeat” of the market — super fast, volatile, driven by buying & selling.

2. Fundamental data = the “health report” of the company — slower moving, rooted in real performance.

Fundamentals give context (is the company strong?), while market data gives signals (is sentiment up or down today?).
